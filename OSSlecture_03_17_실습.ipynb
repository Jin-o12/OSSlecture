```python
# prompt: /content/sample_data/california_housing_train.csv 파일 데이터를 바탕으로 코드를 생성.
# 1. housing_median_age과 total_rooms 데이터 간에 비례 정도를 비교하는 그래프를 만들고 싶음.
# 그래프는 housing_median_age의 수치를 x축으로 두고, 각 값에 대한 평균값을 계산하여, 꺾은 선 그래프로 생성

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# housing_median_age를 기준으로 total_rooms의 평균을 계산합니다.
grouped = df.groupby('housing_median_age')['total_rooms'].mean()

# 꺾은선 그래프를 생성합니다.
plt.figure(figsize=(10, 6))  # 그래프 크기 조절
plt.plot(grouped.index, grouped.values, marker='o', linestyle='-')

# 그래프 제목과 축 레이블을 설정합니다.
plt.title('Housing Median Age vs. Average Total Rooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Average Total Rooms')

# 그리드 추가
plt.grid(True)

# 그래프를 표시합니다.
plt.show()
```
